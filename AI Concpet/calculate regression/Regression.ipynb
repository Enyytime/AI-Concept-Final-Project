```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_csv('D:\\Code\\Coding Stuff\\2nd Year\\AI Concpet\\calculate regression\\gen.csv')


nodes = data['Number of nodes'].values
time = data['Time'].values

plt.scatter(nodes, time, color='blue', label='Data Points')

m, b = np.polyfit(nodes, time, 1) 
plt.plot(nodes, m * nodes + b, color='red', label=f'Linear Fit: y = {m:.4f}x + {b:.4f}')

plt.xlabel('Number of Nodes')
plt.ylabel('Time (minutes)')
plt.title('Scatter Plot and Linear Regression: Time vs. Number of Nodes')

plt.legend()

plt.show()
```
